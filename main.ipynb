melihat perbandingan distribusi TWF secara visual

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='TWF', data=df)
plt.title('Distribusi kegagalan akibat keausan alat')
plt.show()
```

encode label 'Type' dari categorical menjadi numerik

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

le = LabelEncoder()

df['Type'] = le.fit_transform(df['Type'])

X = df.drop(['UDI', 'Product ID', 'TWF', 'Machine failure', 'HDF', 'PWF', 'OSF', 'RNF'], axis=1)
Y = df['TWF']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

melakukkan scaling pada X_train & X_test dan melakukan latihan pada data menggunakan model logisticregression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Buat instance dari StandardScaler
scaler = StandardScaler()

# Fit dan transform data pelatihan
X_train_scaled = scaler.fit_transform(X_train)

# Transform data pengujian
X_test_scaled = scaler.transform(X_test)

# Buat instance dari LogisticRegression
model = LogisticRegression(max_iter=1000)

# Latih model dengan data pelatihan yang telah diskalakan
model.fit(X_train_scaled, Y_train)
```

mencari dan menampilkan predicted TWF

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from matplotlib.colors import LinearSegmentedColormap

Y_pred = model.predict(X_test_scaled)

Y_pred.reshape(-1,1)
```

```python
df_with_prediction = df.copy()

# Mengambil subset dari dataset asli yang sesuai dengan X_test
df_test = df_with_prediction.loc[X_test.index]

# Menambahkan kolom prediksi ke subset dataset asli
df_test["Predicted_TWF"] = Y_pred

# Menampilkan subset dataset dengan kolom prediksi
df_test
```

mencari accuracy score, classification report (precision, recall, dan F1-score)

```python
print("Accuracy:", accuracy_score(Y_test, Y_pred))
print("Classification Report:")
print(classification_report(Y_test, Y_pred, zero_division=1))
```

menampilkan confusion matrix secara visual

```python
cm = confusion_matrix(Y_test, Y_pred)

# Membuat colormap kustom dengan warna darkturquoise
colors = [(0.7, 0.93, 0.93), (0.4, 0.83, 0.86), (0.3, 0.74, 0.84), (0.12, 0.7, 0.66), (0, 0.5, 0.5)]
n_bins = 100  # Discretizes the interpolation into bins
cmap_name = 'my_darkturquoise'
cm_kustom = LinearSegmentedColormap.from_list(cmap_name, colors, N=n_bins)

sns.heatmap(cm, annot=True, fmt="d", cmap= cm_kustom)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

menampilkan visual ROC (Receiver Operating Characteristic)

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Menghitung prediksi probabilitas
Y_pred_prob = model.predict_proba(X_test_scaled)[:, 1]

# Menghitung ROC curve
fpr, tpr, thresholds = roc_curve(Y_test, Y_pred_prob)
roc_auc = roc_auc_score(Y_test, Y_pred_prob)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='limegreen', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='blue', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```
